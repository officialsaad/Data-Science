# 03-How to rename columns

```python
df = pd.DataFrame({'A col': [1, 2,3,4,5,6,7,8,9], 'B col': [3, 4,1,2,5,6,70,8,99]})
```

```python
# rename column and inplace = true means change in the dataset
df.rename(columns={'A col': 'col_a', 'B col': 'col_b'},inplace=True)
df
```

```python
# another way to rename
df.columns = ['col_aab', 'col_b']
df
```

```python
# to replace the _ underscore by char or string
df.columns= df.columns.str.replace('_',' ')
df
```

```python
# adding  prefix 
df = df.add_prefix('BaBa_')
df
```

```python
# add sufix 
df.add_suffix('_BaBa')
print(df)
```

```python
# rename columns    
df.columns= ['cola', 'colb']
df
```
